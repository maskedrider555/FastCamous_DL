# 5. Convert & Insert Data

```python
import datetime
```

```python
# 1 : By declaring a new list as a column
staart_from_101_to_140["요일"] = pd.to_datetime(staart_from_101_to_140["집계일자"], format="%Y%m%d").dt.dayofweek
```

```python
staart_from_101_to_140.head()
```

```python
# 2 : Using .loc[]
staart_from_101_to_140.loc[:, "요일"] = pd.to_datetime(staart_from_101_to_140["집계일자"], format="%Y%m%d").dt.dayofweek
```

```python
staart_from_101_to_140.head()
```

```python
# 3 : Using .asign()
staart_from_101_to_140 = staart_from_101_to_140.assign(요일=pd.to_datetime(staart_from_101_to_140["집계일자"], format="%Y%m%d").dt.dayofweek)
```
