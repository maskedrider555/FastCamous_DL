# 2. Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "C:\DL\FastCampus\DataAnalysis\CSV\data_2021.csv"
data = pd.read_csv(file)
data.info()
```

```python
data_2PM = data[data.집계시 == 14]
data_2PM.info()
```

```python
fig = plt.figure(figsize=(20,10))
```

```python
data_2PM_Destination = sns.countplot("도착영업소코드", data=data_2PM)
data_2PM_Destination.set_title("TransportCount At Destination", fontsize=18)
data_2PM_Destination.set_xlabel("Destination", fontdict={'size':16})
data_2PM_Destination.set_ylabel("TransportCount", fontdict={'size':16})
plt.show(fig)
```

```python
data_2PM_Destination = sns.countplot(y="도착영업소코드", data=data_2PM)
data_2PM_Destination.set_title("TransportCount At Destination", fontsize=18)
data_2PM_Destination.set_xlabel("TransportCount", fontdict={'size':16})
data_2PM_Destination.set_ylabel("Destination", fontdict={'size':16})
plt.show(fig)
```

# 3. Hue Data

```python
data_2PM_Destination = sns.countplot("도착영업소코드", data=data_2PM, hue="요일")
data_2PM_Destination.set_title("TransportCount At Destination", fontsize=18)
data_2PM_Destination.set_xlabel("Destination", fontdict={'size':16})
data_2PM_Destination.set_ylabel("TransportCount", fontdict={'size':16})
plt.show(fig)
```
